```python
import pandas as pd
```

```python
json_path = "/Users/antoine/Desktop/411Project/code/ec_data/schoolInfo.json"
csv_path = "/Users/antoine/Desktop/411Project/code/ec_data/uscities.csv"
df_universities = pd.read_json(json_path)
df_cities = pd.read_csv(csv_path)
```

```python
df_universities
```

```python
df_cities
```

```python
df_merged = pd.merge(df_universities, df_cities, how='left', left_on=['city', 'state'], right_on=['city', 'state_id'])
df_merged.columns 
```

```python
df_merged
```

```python
df_merged = df_merged[['displayName', 'overallRank', 'enrollment', 'sat-avg', 'act-avg', 'city', 'state', 'lat', 'lng']]
```

```python
df_merged = df_merged.rename(columns={
    'displayName': 'University',
    'overallRank': 'Univ_Rank',
    'enrollment': 'Enrollment',
    'sat-avg': 'Avg_SAT',
    'act-avg': 'Avg_ACT',
    'lat': 'Latitude',
    'lng': 'Longitude'
})
df_merged.head()
```

```python
pattern = r'William and Mary|Champaign|Austin|Chapel Hill|in St. Louis|University of Michigan|Texas A&M University--College Station|North Carolina State University--Raleigh'
#MISSING: Santa Clara University, Gonzaga University, Elon University, University of Pittsburgh--Pittsburgh Campus
contains_substrings = df_merged['University'].str.contains(pattern, case=False, na=False)
df_merged[contains_substrings]
```

```python
univ_replacements = {
    'Washington University in St. Louis' : 'Washington University in St Louis',
    'University of Michigan--Ann Arbor' : 'University of Michigan',
    'University of North Carolina--Chapel Hill': 'University of North Carolina at Chapel Hill',
    'College of William and Mary' : 'College of William Mary',
    'University of Illinois--Urbana-Champaign' : 'University of Illinois at Urbana Champaign',
    'University of Texas--Austin' : 'University of Texas at Austin',
    'Texas A&M University--College Station' : 'Texas A M University',
    'North Carolina State University--Raleigh' : 'North Carolina State University'
}
df_merged['University'] = df_merged['University'].replace(univ_replacements)
```

```python
new_universities = [
    'University of Pittsburgh--Pittsburgh Campus',
    'Elon University',
    'Gonzaga University',
    'Santa Clara University'
]
new_rows = pd.DataFrame({
    'University': new_universities,
    'univ_rank': [pd.NA] * len(new_universities),
    'enrollment': [pd.NA] * len(new_universities),
    'avg_sat': [pd.NA] * len(new_universities),
    'avg_act': [pd.NA] * len(new_universities),
    'city': [pd.NA] * len(new_universities),
    'state': [pd.NA] * len(new_universities),
    'latitude': [pd.NA] * len(new_universities),
    'longitude': [pd.NA] * len(new_universities)
})

df_merged = pd.concat([df_merged, new_rows], ignore_index=True)
```

```python
print("Checking for NaNs in the DataFrame:")
nan_columns = df_merged.isnull().any()
print(nan_columns)
print("\nCount of NaNs in each column:")
nan_counts = df_merged.isnull().sum()
print(nan_counts)
print("\nRows with NaNs:")
rows_with_nans = df_merged[df_merged.isnull().any(axis=1)]
print(rows_with_nans)
for column in df_merged.columns:
    if df_merged[column].isnull().any():
        print(f"\nRows with NaNs in column '{column}':")
        print(df_merged[df_merged[column].isnull()])
```

```python
latitude_nan = df_merged['Latitude'].isnull()
longitude_nan = df_merged['Longitude'].isnull()
both_nan = df_merged[latitude_nan & longitude_nan]
print("Entries with both latitude and longitude as NaN:")
print(both_nan)
either_nan = df_merged[latitude_nan | longitude_nan]
print("\nEntries with either latitude or longitude as NaN:")
print(either_nan)
```

```python
#manually add the missing data
manual_lat_long_data = {
    "Dartmouth College": {"Latitude": 43.7044, "Longitude": -72.2887},
    "Boston College": {"Latitude": 42.3355, "Longitude": -71.1685},
    "University of California--San Diego": {"Latitude": 32.8801, "Longitude": -117.2340},
    "Villanova University": {"Latitude": 40.0374, "Longitude": -75.3431},
    "Pennsylvania State University--University Park": {"Latitude": 40.7982, "Longitude": -77.8599},
    "Rutgers University--New Brunswick": {"Latitude": 40.5004, "Longitude": -74.4474},
    "University of New Hampshire": {"Latitude": 43.1336, "Longitude": -70.9320},
    "Clarkson University": {"Latitude": 44.6600, "Longitude": -74.9970},
    "Seton Hall University": {"Latitude": 40.7428, "Longitude": -74.2456},
    "Hofstra University": {"Latitude": 40.7175, "Longitude": -73.6004},
    "Immaculata University": {"Latitude": 40.0272, "Longitude": -75.5957},
    "Maryville University of St. Louis": {"Latitude": 38.6482, "Longitude": -90.5229},
    "Robert Morris University": {"Latitude": 40.5211, "Longitude": -80.2115},
    "University of Hartford": {"Latitude": 41.7981, "Longitude": -72.7149},
    "University of Maine": {"Latitude": 44.8991, "Longitude": -68.6673},
    "Montclair State University": {"Latitude": 40.8646, "Longitude": -74.1986},
    "Nova Southeastern University": {"Latitude": 26.0789, "Longitude": -80.2406},
    "University of Massachusetts--Dartmouth": {"Latitude": 41.6306, "Longitude": -70.9375},
    "Northcentral University": {"Latitude": 33.5855, "Longitude": -112.1252},
    "University of Pittsburgh--Pittsburgh Campus": {"Latitude": 40.4444, "Longitude": -79.9608},
    "Elon University": {"Latitude": 36.1020, "Longitude": -79.5035},
    "Gonzaga University": {"Latitude": 47.6675, "Longitude": -117.4023},
    "Santa Clara University": {"Latitude": 37.3496, "Longitude": -121.9390}
}

def update_lat_long(row):
    university = row['University']
    if university in manual_lat_long_data:
        row['Latitude'] = manual_lat_long_data[university]['Latitude']
        row['Longitude'] = manual_lat_long_data[university]['Longitude']
    return row

df_merged = df_merged.apply(update_lat_long, axis=1)
```

```python
latitude_nan = df_merged['Latitude'].isnull()
longitude_nan = df_merged['Longitude'].isnull()
both_nan = df_merged[latitude_nan & longitude_nan]
print("Entries with both latitude and longitude as NaN:")
print(both_nan)
either_nan = df_merged[latitude_nan | longitude_nan]
print("\nEntries with either latitude or longitude as NaN:")
print(either_nan)
```
